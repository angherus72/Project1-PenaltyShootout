```python

def volado():
    import random
    
    moneda = ''
    choices = ['A','S']
    while moneda != 'A' and moneda != 'S':
        moneda = input('Elige A para Águila o S para Sol ')
        if moneda != 'A' and moneda != 'S':
            print(f'La opción {moneda} no está disponible, por favor selecciona A para Águila y S para Sol')
    choice_volado = random.choice(choices)
    print(choice_volado)
    if moneda == choice_volado:
        print('Felicidades papu, ganaste y tiras primero')
        return 1
    else:
        print('Perdiste paps tiras segundo')
        return 2
```
