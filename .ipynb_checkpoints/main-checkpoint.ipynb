## Función volado

```python

def volado():
    import random
    
    moneda = ''
    choices = ['A','S']
    while moneda != 'A' and moneda != 'S':
        moneda = (input('Elige A para Águila o S para Sol ')).upper()
        if moneda != 'A' and moneda != 'S':
            print(f'La opción {moneda} no está disponible, por favor selecciona A para Águila y S para Sol')
    choice_volado = random.choice(choices)
    print(f'La moneda cayó {choice_volado}')
    if moneda == choice_volado:
        print('Felicidades papu, ganaste y tiras primero')
        return 1
    else:
        print('Perdiste paps tiras segundo')
        return 2
```

## Funciones de tiro

```python
def tiro():
    marcador = 0
    import random
    direccion_tiro = ""
    direcciones_portero = ['C','D','I']
    while direccion_tiro != 'C' and direccion_tiro != 'D' and direccion_tiro != 'I':
        direccion_tiro = (input('Elige C para disparar al Centro, D para disparar a la Derecha e I para disparar a la izquierda ')).upper()
        if direccion_tiro != 'C' and direccion_tiro != 'D' and direccion_tiro != 'I':
            print(f'La opción {direccion_tiro} no es válida papu, por favor elige una opción válida')
    choice_direccion_portero = random.choice(direcciones_portero)
    grafico(direccion_tiro, choice_direccion_portero)
    if direccion_tiro == choice_direccion_portero:
        print('Fallaste el penal papu')
        marcador = 0
    else:
        print('La clavaste papu excelente')
        marcador = 1
        
    return marcador
```

```python
def tirocomputadora():
    import random
    marcador = 0
    direccion_tiro_cpu = ['C','D','I']
    direccion_portero_tu = ""
    while direccion_portero_tu != 'C' and direccion_portero_tu != 'D' and direccion_portero_tu != 'I':
        direccion_portero_tu = (input('Hacia a donde te vas a lanzar papu, D para Derecha, I para izquierda, C para mantenerte en el centro ')).upper()
        if direccion_portero_tu != 'C' and direccion_portero_tu != 'D' and direccion_portero_tu != 'I':
            print(f'La opción {direccion_portero_tu} no es válida papu, por favor elige una opción válida')
    choice_direccion_tiro_cpu = random.choice(direccion_tiro_cpu)
    grafico(choice_direccion_tiro_cpu, direccion_portero_tu)
    if direccion_portero_tu == choice_direccion_tiro_cpu:
        print('Paraste el penal papu felicidades')
        marcador = 0
    else:
        print('Te la clavaron papu, suerte para la otra')
        marcador = 1
    return marcador


```

## Funciones de quienes comienzan tirando

```python
def ganastevolado():
    marcadortu = 0
    marcadorcpu = 0
    recordtirostu = []
    recordtiroscpu = []
    countertu = 0
    countercpu = 0
    tiros = [i for i in range(0,5)]
    for i in tiros:
        if countercpu >= 3:                #pa meterle emocion
            si_anotas_ganas_tu(marcadortu, marcadorcpu, countercpu)
            si_detiene_cpu_pierdes(marcadorcpu, marcadortu, countertu)
        
        if tiro() == 1:
            marcadortu+= 1
            recordtirostu.append('O')
            print(f'JP {marcadortu} CPU {marcadorcpu}')
        else:
            marcadortu+=0
            recordtirostu.append('X')
            print(f'JP {marcadortu} CPU {marcadorcpu}')
        countertu += 1
        if countertu >= 4:
            if criterio_ganador_tu(marcadortu, countertu, marcadorcpu, countercpu) >= 1:
                break

        
        
        if countercpu >= 3:
            si_anotas_cpu_pierdes(marcadorcpu, marcadortu, countertu)
            si_detienes_ganas(marcadortu, marcadorcpu, countercpu)
            
        if tirocomputadora() == 1:
            marcadorcpu += 1
            recordtiroscpu.append('O')
            print(f'JP {marcadortu} CPU {marcadorcpu}')
        else:
            marcadorcpu += 0
            recordtiroscpu.append('X')
            print(f'JP {marcadortu} CPU {marcadorcpu}')
        countercpu += 1
        if countercpu >= 3:                 #esto es lo nuevo
            if criterio_ganador_cpu(marcadorcpu, countercpu, marcadortu, countertu) >= 1:
                break
        
    print(f'El marcador es Jugador: {marcadortu} y CPU: {marcadorcpu}')
    print(f'''Tiros Jugador
    {recordtirostu}
Tiros CPU
    {recordtiroscpu}''')    
    return marcadorcpu, marcadortu, countertu, countercpu
    

```

```python
def perdistevolado():
    marcadortu = 0
    marcadorcpu = 0
    recordtirostu = []
    recordtiroscpu = []
    countertu = 0
    countercpu = 0
    tiros = [i for i in range(0,5)]
    for i in tiros:                           #tiro de la computadora
        if countercpu >= 3:
            si_anotas_cpu_pierdes(marcadorcpu, marcadortu, countertu)
            si_detienes_ganas(marcadortu, marcadorcpu, countercpu)
        
        if tirocomputadora() == 1:
            marcadorcpu+= 1
            recordtiroscpu.append('O')
            print(f'JP {marcadortu} CPU {marcadorcpu}')
        else:
            marcadorcpu+=0
            recordtiroscpu.append('X')
            print(f'JP {marcadortu} CPU {marcadorcpu}')
        countercpu += 1                     # contador de penaltis tirados cpu
        
        if countercpu >= 4:                 #criterio ganador
            if criterio_ganador_cpu(marcadorcpu, countercpu, marcadortu, countertu) >= 1:
                break
        
            
        
        
        if countercpu >= 4:                #pa meterle emocion
            si_anotas_ganas_tu(marcadortu, marcadorcpu, countercpu)
            si_detiene_cpu_pierdes(marcadorcpu, marcadortu, countertu)
            
        if tiro() == 1:                     #tiro tu
            marcadortu += 1
            recordtirostu.append('O')
            print(f'JP {marcadortu} CPU {marcadorcpu}')
        else:
            marcadortu += 0
            recordtirostu.append('X')
            print(f'JP {marcadortu} CPU {marcadorcpu}')
        countertu += 1                      # contador de penaltis tirados tu
        
        if countertu >= 3:                  #criterio ganador
            if criterio_ganador_tu(marcadortu, countertu, marcadorcpu, countercpu) >= 1:
                break
        
        

    print(f'El marcador es Jugador: {marcadortu} y CPU: {marcadorcpu}')
    print(f'''Tiros Jugador
    {recordtirostu}
Tiros CPU
    {recordtiroscpu}''')
    
    return marcadorcpu, marcadortu, countertu, countercpu
    

```

## Criterios ganadores 

```python
def criterio_ganador_cpu(marcador1,counter1,marcador2,counter2):
    penales_restantes1 = 5 - counter1 
    penales_restantes2 = 5 - counter2

    if marcador1 + penales_restantes1 < marcador2:
        print('Haz ganado papu felicidades')
        return 1
    elif marcador1 > marcador2 + penales_restantes2:
        print('Haz perdido')
        return 2
    else:
        return 0


```

```python
def criterio_ganador_tu(marcador1,counter1,marcador2,counter2):
    penales_restantes1 = 5 - counter1 
    penales_restantes2 = 5 - counter2

    if marcador1 > marcador2 + penales_restantes2:
        print('Haz ganado papu felicidades')
        return 1
    elif marcador1 + penales_restantes1 < marcador2:
        print('CPU Ha ganado')
        return 2
    else:
        return 0
```

## Funciones para avivar el juego

```python
def si_anotas_ganas_tu(marcador1,marcador2,counter2): 
    penales_restantes2 = 5 - counter2
    
    if marcador1 + 1 > marcador2 + penales_restantes2:
        print('Si metes el siguiente penal ganas')
        return 1
    
    
```

```python
def si_anotas_cpu_pierdes(marcador1,marcador2,counter2): 
    penales_restantes2 = 5 - counter2
    
    if marcador1 + 1 > marcador2 + penales_restantes2:
        print('Si la CPU mete el siguiente penal pierdes')
        return 1
    
    
```

```python
def si_detienes_ganas(marcador1,marcador2,counter2):
    penales_restantes2= 5-counter2
    
    if marcador1 > marcador2 + penales_restantes2 - 1:
        print('Si paras el siguiente penal ganas')
        return 1
```

```python
def si_detiene_cpu_pierdes(marcador1,marcador2,counter2): 
    penales_restantes2 = 5 - counter2
    
    if marcador1 > marcador2 + penales_restantes2 - 1:
        print('Si la CPU para el siguiente penal pierdes')
        return 1
    
```
